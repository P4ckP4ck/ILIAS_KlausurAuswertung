Theta = Maß für Fähigkeit des Studierenden

a = Diskriminationsvariable, wie schnell steigt die Lernkurve

b = Schwierigkeit der Frage

c = pseudo-guessing Wahrscheinlichkeit

Wie hoch ist die Wahrscheinlichkeit, dass jemand die Frage richtig beantwortet, obwohl sie nicht seinen Fähigkeiten entspricht


d = Unaufmersamkeitsfaktor \n

Wie hoch ist die Wahrscheinlichkeit, dass jemand die Frage falsch beantwortet, obwohl sie seinen Fähigkeiten entspricht.
Der Wert entspricht 1 minus der Warscheinlichkeit des Fehlers!!
